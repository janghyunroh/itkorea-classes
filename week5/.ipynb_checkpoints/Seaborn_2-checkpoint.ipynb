```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')

plt.rcParams['font.family'] = 'Noto Serif CJK JP'
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 타이타닉 데이터 불러오기
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# 스타일 테마 설정
sns.set_style('darkgrid')
```

```python
#그래프 객체 생성
fig = plt.figure(figsize=(15, 5))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

# seaborn : 대학에서 통계 목적으로 만듦
# 논문 등 학문적, 전문적 통계 표시에 강함

#regplot : plot의 scatter와 같음, 회귀선 + 신뢰구간이 기본으로 설정됨
#fir_reg 옵션으로 끌 수 있음
sns.regplot(x = 'age',
            y = 'fare',
            data=titanic,
            ax=ax1)

sns.regplot(x = 'age',
            y = 'fare',
            data=titanic,
            ax=ax2,
           fit_reg = False)
```

```python
#dist, kde, hist
#distplot : 히스토그램 + 라인 그래프
#kdeplot : 라인 그래프만 - density
#histplot: 히스토그램만 - count

fig = plt.figure(figsize=(15, 5))
ax1 = fig.add_subplot(1, 3, 1)
ax2 = fig.add_subplot(1, 3, 2)
ax3 = fig.add_subplot(1, 3, 3)

#ax1 distplot
sns.distplot(titanic['fare'], ax=ax1) # 방법 1

sns.kdeplot(x='fare', data=titanic, ax=ax2) # 방법 2

sns.histplot(x='fare', data=titanic, ax=ax3)
```

```python
# 피벗 테이블
# groupby : 컬럼 별 계산 적용
# 피벗 테이블 : 컬럼 별 묶기 (함수 통해 추가 계산 가능)

table = titanic.pivot_table(index=['sex'], columns=['class'], aggfunc='size')
table
```

```python
## 히트맵 : 피벗 테이블 시각화
sns.heatmap(table, 
            annot=True,fmt='d',
            cmap='YlGnBu',
            linewidth=.5,
            cbar=True)
plt.show()
```

```python
# 상관계수 떠올리기!
titanic.corr(numeric_only=True)
```

```python
sns.heatmap(titanic.corr(numeric_only=True), 
            annot=True,
            cmap='YlGnBu',
            linewidth=.5,
            cbar=True)
plt.show()
```

```python
# scatter : 상관관계 + 분포도까지
# heatmap : 상관계수의 정확한 수치 표현
```

```python
# scatterplt은 두 데이터 모두 이산형이어야 했음
# stripplot, swarmpplot
# 한 축은 범주 한 축은 이산형이어야 함


fig = plt.figure(figsize=(15, 5))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

#ax1 : stripplit
sns.stripplot(x='class',
              y='age',
              data=titanic,
              ax=ax1)

sns.swarmplot(x='class',
              y='age',
              data=titanic,
              ax=ax2)
# 분산을 고려해서 많이 뭉쳐있는 곳을 시각적으로 뚱뚱하게 만들어 분포를 더 잘 보이게 함

#violinplot은 나중에 배우기!
plt.show()
```

```python
# barplot
fig = plt.figure(figsize=(15, 5))
ax1 = fig.add_subplot(1, 3, 1)
ax2 = fig.add_subplot(1, 3, 2)
ax3 = fig.add_subplot(1, 3, 3)

# count 
sns.barplot(x='sex', y='survived', data=titanic, ax=ax1)

# hue 옵션 : 추가적으로 나눌 standard 지정
sns.barplot(x='sex', y='survived', hue='class', data=titanic, ax=ax2) # 안에서 추가적으로 나눌 standard(범주형이어야 함)

# dodge 옵션 : 히스토그램 간 안 겹치기 -> 끄면 hue로 분리한 게 겹쳐서 나옴
# -> 차이를 육안으로 볼 수 있음!
sns.barplot(x='sex', y='survived', hue='class', dodge=False, data=titanic, ax=ax3)
```

```python
# countplot - 개수 세주기
fig = plt.figure(figsize=(15, 5))
ax1 = fig.add_subplot(1, 3, 1)
ax2 = fig.add_subplot(1, 3, 2)
ax3 = fig.add_subplot(1, 3, 3)

#기본 그래프
sns.countplot(x='class', palette='Set1', data=titanic, ax=ax1)

# hue 옵션
sns.countplot(x='class', palette='Set2',hue='who', data=titanic, ax=ax2)

sns.countplot(x='class', palette='Set3',hue='who', dodge=False, data=titanic, ax=ax3)
```

```python
# 마지막 그래프 - joint 그래프
sns.set_style('whitegrid')

# 조인트 그래프 - 기본값(산점도)
j1 = sns.jointplot(x='fare', y='age', data=titanic)

# 회귀선 
j2 = sns.jointplot(x='fare', y='age', kind='reg', data=titanic)

# 육각 그래프
j3 = sns.jointplot(x='fare', y='age', kind='hex', data=titanic)
# 점을 무작정 찍는 것이 아니라 색깔 있는 육각점으로 분포를 더 정확히 표현!
```
