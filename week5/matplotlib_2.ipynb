```python
# 라이브러리
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

plt.rcParams['font.family'] = 'Noto Serif CJK JP'
plt.rcParams['axes.unicode_minus'] = False
```

```python
path = '~/share/시도별 전출입 인구수.xlsx'
df = pd.read_excel(path, engine='openpyxl')
df.head()
```

```python
df = df.fillna(method='ffill')
```

```python
# 서울 -> 다른 지역
mask = (df['전출지별']=='서울특별시') & (df['전입지별'] != '서울특별시')
df_seoul = df[mask]
df_seoul = df_seoul.drop(['전출지별'], axis =1)
df_seoul.rename({'전입지별': '전입지'}, axis = 1, inplace = True)
df_seoul.set_index('전입지', inplace = True)

df_seoul.head()
```

# 내가 원하는 그래프의 x축이 인덱스가 되도록 하는 것이 중요! -> Transpose

목표 그래프

x축: 연도

y축: 인구 이동량

데이터 종류: 서울 -> 충남, 서울 -> 경북, 서울 -> 강원, 서울 -> 전남

```python
#column list 만들기
col_years = list(map(str, range(1970, 2018)))

#서울 -> 충남, 경북, 강원, 전남 데이터만 가져오기
df_4 = df_seoul.loc[['충청남도', '경상북도', '강원도', '전라남도']]
df_4 = df_4.T
```

```python
#면적 그래프 그려보기
plt.style.use('ggplot') # - 논문 쓸 때 가장 많이 쓰는 스타일임!(R에서부터 계승되온 전통)

#면적 그래프 객체 생성
df_4.plot(kind = 'area', stacked=False, alpha=0.2, figsize = (20,10))
# kind: 
# stacked: 누적 
# alpha: 투명도
# figsize: 

plt.title('서울 -> 타도시 인구 이동', size = 20)
plt.ylabel('이동 인구 수', size = 15)
plt.xlabel('기간', size = 15)
plt.legend(loc='best', fontsize = 13)

plt.show()

# 면적 그래프에 stacked=False로 두고 쓸 때는 주로 비율을 보고 싶을 때 사용! 
```

```python
plt.style.use('ggplot') # - 논문 쓸 때 가장 많이 쓰는 스타일임!(R에서부터 계승되온 전통)

#면적 그래프 객체 생성
df_4.plot(kind = 'area', stacked=True, alpha=0.2, figsize = (20,10))
# kind: 
# stacked: 누적 
# alpha: 투명도
# figsize: 

plt.title('서울 -> 타도시 인구 이동', size = 20)
plt.ylabel('이동 인구 수', size = 15)
plt.xlabel('기간', size = 15)
plt.legend(loc='best', fontsize = 13)

plt.show()
```

```python
# 막대 그래프 그리기
df_4.plot(kind='bar', figsize = (20, 10), width=0.7, color = ['orange', 'green', 'skyblue', 'blue'])

#color는 이렇게 리스트로 줄 수도 있음!
# 그리고 color name이 아니라 RGB 코드로 줄 수도 있음!
plt.title('서울 -> 타도시 인구 이동', size = 20)
plt.ylabel('이동 인구 수', size = 15)
plt.xlabel('기간', size = 15)
plt.legend(loc='best', fontsize = 13)

plt.show()
```

```python
# 문제점: 보여줄 데이터가 너무 많아 지저분함!
# 막대 그래프로 그려야 한다면 어떻게 할까?
# 1) 보고 싶은 부분만 잘라서 보여주기
# 2) 평균 내기

# 서울 -> 충남, 경북, 강원, 전남의 연도별 평균 인구 이동 그래프 그리기!
sr_mean = df_4.mean(axis=1)

sr_mean.plot(kind='bar', figsize = (20, 10), width=0.7, color = 'orange')

#color는 이렇게 리스트로 줄 수도 있음!
# 그리고 color name이 아니라 RGB 코드로 줄 수도 있음!
plt.title('서울 -> 타도시 인구 이동 평균', size = 20)
plt.ylabel('이동 인구 수', size = 15)
plt.xlabel('기간', size = 15)


plt.show()
```

```python
# 다른 방법 - 코드 한 줄만 추가해서?
# 막대 그래프 그리기
df_4 = df_4.mean(axis=1)
df_4.plot(kind='bar', figsize = (20, 10), width=0.7, color = ['orange', 'green', 'skyblue', 'blue'])

#color는 이렇게 리스트로 줄 수도 있음!
# 그리고 color name이 아니라 RGB 코드로 줄 수도 있음!
plt.title('서울 -> 타도시 인구 이동', size = 20)
plt.ylabel('이동 인구 수', size = 15)
plt.xlabel('기간', size = 15)
plt.legend(loc='best', fontsize = 13)

plt.show()
```

```python
# 다른 방법 - 평균 칼럼을 만들고 그리기
df_4['평균'] = df_4.mean(axis=1)
plt.figure(figsize = (20, 10))
plt.bar(col_years, df_4['평균'], color='blue')
plt.show()
```

```python
# 보조축 그리기
# y값이 다른 데이터보다 크게 달라서 별도의 y축을 두는게 나은 경우 - 보조축 활용!
df = pd.read_excel('~/share/electric.xlsx', engine = 'openpyxl')
df
```

```python
# 북한 데이터만 가져오기
df = df.loc[5:9]
df
```

```python
df.drop('전력량 (억㎾h)', axis = 1, inplace = True)
df.set_index('발전 전력별', inplace = True)
df = df.T
df
```

```python
df = df.rename(columns={'합계':'총발전량'})
df['작년 총발전량'] = df['총발전량'].shift(1) # shift함수 - 해당 인덱스의 바로 전 인덱스 데이터 가져옴
df
```

```python
df['증감율'] = ( (df['총발전량']/df['작년 총발전량']) - 1) * 100
df
```

```python
#이제 드디어 그래프 그리자!
#목표 그래프: 북한의 수력, 화력 발전량의 총합 및 전년도 대비 발전 증감률
plt.style.use('ggplot')

#2축 그래프 생성하고 그리기

#1. 수력, 화력 발전량 합
ax1 = df[['수력', '화력']].plot(kind='bar', figsize=(20, 10), width=0.8, stacked=True)

#2. 전년 대비 발전량 증감률
ax2 = ax1.twinx()    # twinx function
ax2.plot(df.index, df['증감율'], ls='--', marker='o', markersize=15,
        color='green', label='전년대비 증감율(%)')
#ls : line style

#여기까지 하면 굉장히 보기 안좋음 -> 막대 그래프랑 라인 그래프랑 너무 겹쳐있다보니 
# --> 

ax1.set_ylim(0, 500)
ax2.set_ylim(-50, 50)

# 데이터 분석 기사 시험 때는 limit 값 고려하기!

ax1.set_xlabel('연도', size=20)
ax1.set_ylabel('발전량(억 KWh)')

ax2.set_ylabel('전년 대비 증감율(%)')

ax1.set_title('북한 전력 발전량 (1990 ~ 2016)', size=30)
ax1.legend(loc='upper left')

plt.show()
```
