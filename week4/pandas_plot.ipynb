```python
# read_csv auto-mpg
df = pd.read_csv('~/share/auto-mpg.csv', header=None)
df.columns = ['연비', '실린더 수', '배기량', '출력', '차종', '가속능력', '출시년도', '제조국', '모델명']
df.head()
```

```python
# scatter(산점도) : 두 데이터 간의 상관관계 시각화
df.plot(x = '차종', y = '연비', kind='scatter')
```

```python
df.plot(x='차종', y='실린더 수', kind='scatter')
```
