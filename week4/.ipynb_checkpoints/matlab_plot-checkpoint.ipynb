```python
# 라이브러리
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

plt.rcParams['font.family'] = 'Noto Serif CJK JP'
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('~/share/시도별 전출입 인구수.xlsx', engine='openpyxl')
```

```python
df.head(30)
```

```python
# 누락값 채우기
df = df.fillna(method='ffill')
df.head(30)
```

```python
### 서울에서 다른 지역으로 이동한 데이터만 추출하여 df_seoul을 만드시오.
mask = (df['전출지별']=='서울특별시') & (df['전입지별']!='서울특별시')
df_seoul = df[mask]
df_seoul = df_seoul.drop(['전출지별'], axis=1)
df_seoul.rename({'전입지별':'전입지'}, axis=1, inplace=True)
df_seoul.set_index('전입지', inplace=True)
```

```python
df_seoul
```

```python
# 서울에서 경기도로 이주한 인구 데이터만 가져오기
sr_one = df_seoul.loc['경기도']
sr_one
```

```python
# matplot 그래프 그리기
plt.plot(sr_one.index, sr_one.values)
```

```python
# 도화지의 크기를 결정
plt.figure(figsize=(15, 5))

# 스타일 가이드 설정
plt.style.use('ggplot')

plt.plot(sr_one.index, sr_one.values, marker = 'o', markersize=5)

plt.title('서울 -> 경기 인구 이동', size=15)

plt.xlabel('기간', size=10)
plt.xticks(size=8, rotation='vertical')
plt.ylabel('이동 인구수', size=10)

plt.legend(labels=['서울 -> 경기'], loc='best', fontsize=10)

plt.show()
```

```python
# annotation 달기

#1. size
plt.figure(figsize=(15,5))
#2. style
plt.style.use('ggplot')
#3. y축 범위 설정
plt.ylim(50000, 800000)

plt.plot(sr_one.index, sr_one.values, marker='o', markersize=5)

plt.title('서울->경기 인구 이동', size=15)
plt.xlabel('기간', size=10)
plt.xticks(size=8, rotation='vertical')
plt.ylabel('이동 인구수', size=10)
plt.legend(labels=['서울->경기'], loc='best', fontsize=12)


##
plt.annotate('', 
             xy= (sr_one.index.get_loc('1986'), 620000), 
             xytext=(2, 290000),
             arrowprops=dict(arrowstyle='->', color='skyblue', lw = 5)
            
            )
plt.annotate('',
            xy = (sr_one.index.get_loc('2014'), 430000),   # head 좌표
            xytext=(sr_one.index.get_loc('2000'), 620000), # tail 좌표
             arrowprops=dict(arrowstyle='->', color='red', lw=5)
            
            
            )


## text 작섣
plt.annotate('인구 이동 증가(1979-1995)',
             xy=(8, 400000),
             ha='center',
             fontsize=15,
             rotation=29)

plt.annotate('인구 이동 감소(1995-2017)',
             xy=(37, 500000),
             ha='center',
             fontsize=15,
             rotation=-15)

plt.show()
```

```python
#서울에서 충청남도, 경상북도, 강원도
df_3 = df_seoul.loc[['충청남도', '경상북도', '강원도']]
col_years = list(map(str, range(1970, 2018)))
```

```python
df_3
```

```python
fig = plt.figure(figsize=(20, 5))
ax = fig.add_subplot(1, 1, 1)

ax.plot(col_years, df_3.loc['충청남도', :], marker='o', markerfacecolor='green',
        markersize=5,color='olive', linewidth=2, label='서울 -> 충남')

ax.plot(col_years, df_3.loc['경상북도', :], marker='o', markerfacecolor='blue',
        markersize=5,color='skyblue', linewidth=2, label='서울 -> 경북')

ax.plot(col_years, df_3.loc['강원도', :], marker='o', markerfacecolor='red',
        markersize=5,color='magenta', linewidth=2, label='서울 -> 강원')

ax.legend(loc='best')

ax.set_title('서울 -> 충남, 경북, 강원 인구 이동', size=15)

ax.set_xlabel('기간', size=10)

ax.set_ylabel('이동 인구수', size=10)

ax.set_xticklabels(col_years, rotation=90)

ax.tick_params(axis='x', labelsize=8)
ax.tick_params(axis='y', labelsize = 8)
plt.show()
```

# dataframe
충남 -> 서울  제주 -> 전국
전남 -> 부산  서울 -> 부산

```python
#1. 충남 -> 서울
mask_chung = (df['전출지별']=='충청남도') & (df['전입지별']!='충청남도')
df_chung = df[mask_chung]
df_chung = df_chung.drop(['전출지별'], axis=1)
df_chung.rename({'전입지별':'전입지'}, axis=1, inplace=True)
df_chung.set_index('전입지', inplace=True)
#
sr_cs = df_chung.loc['서울특별시']
```

```python
#2. 제주 -> 전국
mask_jeju = (df['전출지별']=='제주특별자치도') & (df['전입지별']!='제주특별자치도')
df_jeju = df[mask_jeju]
df_jeju = df_jeju.drop(['전출지별'], axis=1)
df_jeju.rename({'전입지별':'전입지'}, axis=1, inplace=True)
df_jeju.set_index('전입지', inplace=True)
#
sr_jeju = df_jeju.loc['전국']
```

```python
#3. 전남 -> 서울
mask_jun = (df['전출지별']=='전라남도') & (df['전입지별']!='전라남도')
df_jun = df[mask_jun]
df_jun = df_jun.drop(['전출지별'], axis=1)
df_jun.rename({'전입지별':'전입지'}, axis=1, inplace=True)
df_jun.set_index('전입지', inplace=True)
#
sr_jun = df_jun.loc['부산광역시']
```

```python
#4. 서울 -> 부산
mask_seoul = (df['전출지별']=='서울특별시') & (df['전입지별']!='서울특별시')
df_seoul = df[mask_seoul]
df_seoul = df_seoul.drop(['전출지별'], axis=1)
df_seoul.rename({'전입지별':'전입지'}, axis=1, inplace=True)
df_seoul.set_index('전입지', inplace=True)
#
sr_seoul = df_seoul.loc['부산광역시']
```

```python
#그래프 객체 생성
fig = plt.figure(figsize=(20, 5))

#
ax_chun = fig.add_subplot(2, 2, 1)
ax_chun.plot(col_years, sr_cs, marker='o', markersize=5, markerfacecolor='red', color='magenta', linewidth=2, label='충남 -> 서울')
ax_chun.legend(loc='best')
ax_chun.set_title('충남 -> 서울 인구 이동', size=15)
ax_chun.set_xlabel('기간', size=10)
ax_chun.set_ylabel('이동 인구수', size=10)
ax_chun.set_xticklabels(col_years, rotation=90)

ax_chun.tick_params(axis='x', labelsize=8)
ax_chun.tick_params(axis='y', labelsize = 8)

#
ax_jeju = fig.add_subplot(2, 2, 2)
ax_jeju.plot(col_years, sr_jeju, marker='o', markersize=5, markerfacecolor='green', color='olive', linewidth=2, label='제주 -> 전국')
ax_jeju.legend(loc='best')
ax_jeju.set_title('제주 -> 전국 인구 이동', size=15)
ax_jeju.set_xlabel('기간', size=10)
ax_jeju.set_ylabel('이동 인구수', size=10)
ax_jeju.set_xticklabels(col_years, rotation=90)

ax_jeju.tick_params(axis='x', labelsize=8)
ax_jeju.tick_params(axis='y', labelsize = 8)
#
ax_jun = fig.add_subplot(2, 2, 3)
ax_jun.plot(col_years, sr_jun, marker='o', markersize=5, markerfacecolor='green', color='olive', linewidth=2, label='전남 -> 부산')
ax_jun.legend(loc='best')
ax_jun.set_title('전남 -> 부산 인구 이동', size=15)
ax_jun.set_xlabel('기간', size=10)
ax_jun.set_ylabel('이동 인구수', size=10)
ax_jun.set_xticklabels(col_years, rotation=90)

ax_jun.tick_params(axis='x', labelsize=8)
ax_jun.tick_params(axis='y', labelsize = 8)

#
ax_seoul = fig.add_subplot(2, 2, 4)
ax_seoul.plot(col_years, sr_seoul, marker='o', markersize=5, markerfacecolor='green', color='olive', linewidth=2, label='서울 -> 부산')
ax_seoul.legend(loc='best')
ax_seoul.set_title('서울 -> 부산 인구 이동', size=15)
ax_seoul.set_xlabel('기간', size=10)
ax_seoul.set_ylabel('이동 인구수', size=10)
ax_seoul.set_xticklabels(col_years, rotation=90)

ax_seoul.tick_params(axis='x', labelsize=8)
ax_seoul.tick_params(axis='y', labelsize = 8)

plt.show()
```
