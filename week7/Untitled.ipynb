# 프로젝트 

Kaggle Dataset과 KOSIS MicroDataset을 사용하여 데이터 분석!

## 1. 주제 선정(목적 설정)

내가 어떤 데이터를 다룰 것인가?(도메인을 자세히 결정)

10:30까지 정하기

e.g. 영화 산업에서의 마케팅

## 2. 데이터 수집

주제에 맞는 필요한 데이터 수집

kaggle이나 아니면 다른 곳에서 수집

e.g. 최근 개봉 영화의 관객 수


## 3. 데이터 분석

### 1) 가설 설정
목표를 위해 5가지 정도 의 가설을 설정

e.g. 히어로물은 10~20대에게 가장 인기가 많을 것이다.

e.g. 히어로 중 가장 인기가 많은 유형은 속도가 빠른 히어로이다.

### 2) 가설 증명 및 시각화 (EDA)

CDA 말고 EDA(시각화) 하기!

## 4. 분석 결과를 기반으로 한 의사 결정 

증빙하기 위한 자료로 추가 세부 가설 등을 추가로 설정

e.g. 10~20대를 위한 프로모션 등은 어떤 것이 좋을지

## 5. 주의사항

### 1) 
가설이 데이터에 적합하지 않은 경우 (가설이 틀린 경우) 가설을 버리는 게 아니라 

얻을 수 있는 추가적인 insight를 이용해 가설을 새로 세울 것!

### 2) 
가설의 증명이 허점이 있으면 안됨!

### 3)
사견이 포함되면 안됨!

### 4) 
거짓말 하지 말 것!

# 2. 가설 설정

1) 운전자의 연령이 높을수록 사고 발생률이 증가할 것이다. 따라서 노령 운전자에 대한 운전 보수 교육, 또는 특정 연령을 넘어가면 운전자 면허 반납

    [세부 가설 및 참고 기사]

    1 - 1) 운전 보수 교육을 마친 운전자에 대한 사고 발생률 추이

2) 운전자의 운전 경력이 짧을수록 사고 발생률이 증가할 것이다. 

3) 하루 24시간 중 출퇴근 시간에 사고 발생률이 가장 높을 것이다. 

4) 1년 365일 중 연휴 기간에 사고 발생률이 가장 높을 것이다. 

5) 여름보다 겨울에 더 사고 발생률이 높을 것이다. 

6) 교통량이 많은 업무지구가 주거지구보다 사고 발생률이 더 높을 것이다.
